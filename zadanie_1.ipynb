```python
import openpyxl
import pandas as pd
import numpy as np
from dataclasses import dataclass
from datetime import date
from statistics import NormalDist
import matplotlib.pyplot as plt
from lib.dates import year_fraction
from lib.import_data import load_market, MarketEngine
from lib.vanilla_option import OptionType, VanillaOptionBlackSholes, d,strike_for_delta, vanna_volga_cost_coefficients
from lib.binary_option import CoNBlackScholes, AoNBlackScholes
from lib.digital_option import OneTouchBlackScholes, NoTouchBlackScholes, PaymentType, BoundType, theta, vartheta, d as d_digi

class PricingEngine: #główny silnik odpowiedzialny za wycenę,
    def __init__(self, market_file: str, tenor: str, imply_pln: bool): #ładowanie danych, obliczanie stóp, zmienności ect.
        self.market = load_market(market_file, tenor)
        self.m_engine = MarketEngine(self.market)

        self.r_pln, self.r_eur, self.df_d, self.df_f = (self.m_engine.rates_dfs(imply_pln))

        self.sigma_atm = self.market.atm
        self.sigma_25C = self.market.atm + self.market.bf25 + 0.5 * self.market.rr25
        self.sigma_25P = self.market.atm + self.market.bf25 - 0.5 * self.market.rr25
        
    def strike(self,delta: float,option_type: OptionType): #funkcja znajdująca strike funkcji dla sigma ATM.
        sigma = self.sigma_atm
        return strike_for_delta(
            delta=delta,
            option_type=option_type,
            df_d=self.df_d,
            df_f=self.df_f,
            S_t=self.market.spot,
            sigma=sigma,
            t=self.market.start,
            T=self.market.expiry,
            base=365,
            forward=self.market.delta_forward,
            premium=self.market.delta_premium,
        )
    #w poniższej funkcji AI pomogło refaktoryzować kod w celu lepszej przejrzystości i aby łatwiej było powielać schemat wycen dla innych typów
    def vanilla_price(self,K: float,option_type: OptionType,pricing_method: str,p: float = 1.0,): #wycena opcji waniliowej wg. BS i VV
        option = VanillaOptionBlackSholes(T=self.market.expiry,K=K,option_type=option_type)
        
        if pricing_method.lower() == "bs":
            return option.price(
                df_d=self.df_d,
                df_f=self.df_f,
                S_t=self.market.spot,
                sigma=self.sigma_atm,
                t=self.market.start,
                base=365,
            )

        elif pricing_method.lower() == "vv":
            return option.vanna_volga_price(
                df_d=self.df_d,
                df_f=self.df_f,
                p=p,
                S_t=self.market.spot,
                sigma_atm=self.sigma_atm,
                sigma_25C=self.sigma_25C,
                sigma_25P=self.sigma_25P,
                t=self.market.start,
                delta_forward=self.market.delta_forward,
                delta_premium=self.market.delta_premium,
                base=365,
            )

        raise ValueError("Metoda to nie BS ani VV")

    def con_price(self,K: float,option_type: OptionType,pricing_method: str = "bs",p: float = 1.0,dK: float = 0.0001,): #wycena opcji CoN wg. BS i VV
        option = CoNBlackScholes(T=self.market.expiry, K=K, option_type=option_type )
        if pricing_method.lower() == "bs":
            return option.price(
                df_d=self.df_d,
                df_f=self.df_f,
                S_t=self.market.spot,
                sigma=self.sigma_atm,
                t=self.market.start,
                base=365,
            )
        elif pricing_method.lower() == "vv":
            return option.vanna_volga_price(
                df_d=self.df_d,
                df_f=self.df_f,
                p=p,
                S_t=self.market.spot,
                sigma_atm=self.sigma_atm,
                sigma_25C=self.sigma_25C,
                sigma_25P=self.sigma_25P,
                t=self.market.start,
                delta_forward=self.market.delta_forward,
                delta_premium=self.market.delta_premium,
                base=365,
                dK=dK,
            )

        raise ValueError("Metoda to nie BS ani VV")

    def aon_price(self,K: float,option_type: OptionType,pricing_method: str = "bs",p: float = 1.0,dK: float = 0.0001): #wycena opcji AoN wg. BS i VV
        option = AoNBlackScholes(T=self.market.expiry,K=K,option_type=option_type)
        
        if pricing_method.lower() == "bs":
            return option.price(
                df_d=self.df_d,
                df_f=self.df_f,
                S_t=self.market.spot,
                sigma=self.sigma_atm,
                t=self.market.start,
                base=365,
            )

        elif pricing_method.lower() == "vv":
            return option.vanna_volga_price(
                df_d=self.df_d,
                df_f=self.df_f,
                p=p,
                S_t=self.market.spot,
                sigma_atm=self.sigma_atm,
                sigma_25C=self.sigma_25C,
                sigma_25P=self.sigma_25P,
                t=self.market.start,
                delta_forward=self.market.delta_forward,
                delta_premium=self.market.delta_premium,
                base=365,
                dK=dK,
            )

        raise ValueError("Metoda to nie BS ani VV")

    def one_touch_price(self,B :float ,option_type: OptionType, bound_type: BoundType, payment_type: PaymentType, pricing_method: str = "bs",p: float = 1.0):
        option = OneTouchBlackScholes(T=self.market.expiry,B=B,bound_type=bound_type, payment_type=payment_type) #wycena opcji OneTouch wg. BS i VV
        
        if pricing_method.lower() == "bs":
            return option.price(
                df_d=self.df_d,
                df_f=self.df_f,
                S_t=self.market.spot,
                sigma=self.sigma_atm,
                t=self.market.start,
                base=365,
            )

        elif pricing_method.lower() == "vv":
            return option.vanna_volga_price(
                df_d=self.df_d,
                df_f=self.df_f,
                p=p,
                S_t=self.market.spot,
                sigma_atm=self.sigma_atm,
                sigma_25C=self.sigma_25C,
                sigma_25P=self.sigma_25P,
                t=self.market.start,
                delta_forward=self.market.delta_forward,
                delta_premium=self.market.delta_premium,
                base=365,
            )
    def no_touch_price(self,B :float ,option_type: OptionType, bound_type: BoundType, pricing_method: str = "bs",p: float = 1.0):
        option = NoTouchBlackScholes(T=self.market.expiry,B=B,bound_type=bound_type)        #wycena opcji NoTouch wg. BS i VV
        
        if pricing_method.lower() == "bs":
            return option.price(
                df_d=self.df_d,
                df_f=self.df_f,
                S_t=self.market.spot,
                sigma=self.sigma_atm,
                t=self.market.start,
                base=365,
            )

        elif pricing_method.lower() == "vv":
            return option.vanna_volga_price(
                df_d=self.df_d,
                df_f=self.df_f,
                p=p,
                S_t=self.market.spot,
                sigma_atm=self.sigma_atm,
                sigma_25C=self.sigma_25C,
                sigma_25P=self.sigma_25P,
                t=self.market.start,
                delta_forward=self.market.delta_forward,
                delta_premium=self.market.delta_premium,
                base=365,
            )

    # Poniżej jest funkcja, której zadaniem jest obliczenie wartości opcji w zależności od Spotu według BS i VV.
    # Zawiera paramatery dla wszystkich opcji, jednak tylko część z nich jest używana, n oznacza gęstość obliczenia wartości,
    #zaś b to jak szeroki jest przedział liczenia
    def option_price_range(self, product_type:str, option_type: OptionType, K: float, pricing_method: str,  
                           bound_type: BoundType,payment_type: PaymentType, n: int = 100, b: float=0.05):
        S0 = self.market.spot
        spots = np.linspace(S0 *(1-b), S0 *(1+b), n)
        prices = []

        match product_type: #wybór typu opcji
            case "Vanilla":
                option = VanillaOptionBlackSholes(T=self.market.expiry,K=K,option_type=option_type)
            case "CoN":
                option = CoNBlackScholes(T=self.market.expiry,K=K,option_type=option_type)
            case "AoN":
                option = AoNBlackScholes(T=self.market.expiry,K=K,option_type=option_type)
            case "OneTouch":
                option = OneTouchBlackScholes(T=self.market.expiry,B=K,bound_type=bound_type,payment_type=payment_type)     
            case "NoTouch":
                option = NoTouchBlackScholes(T=self.market.expiry,B=K,bound_type=bound_type) 
            case _:
                raise ValueError(f"Nieznany typ opcji: {product_type}")
        for S in spots: #wyliczanie wartości opcji wg bs i vv
            
            if pricing_method=="bs":
                price = option.price(
                df_d=self.df_d,
                df_f=self.df_f,
                S_t=S,
                sigma=self.sigma_atm,
                t=self.market.start,
                base=365,
                )
            elif pricing_method == "vv":
                if product_type=="Vanilla" or product_type=="OneTouch" or product_type=="NoTouch" :
                    price = option.vanna_volga_price(
                    df_d=self.df_d,
                    df_f=self.df_f,
                    p=1.0,
                    S_t=S,
                    sigma_atm=self.sigma_atm,
                    sigma_25C=self.sigma_25C,
                    sigma_25P=self.sigma_25P,
                    t=self.market.start,
                    delta_forward=self.market.delta_forward,
                    delta_premium=self.market.delta_premium,
                    base=365,)
                elif product_type=="CoN" or product_type=="AoN":
                    price = option.vanna_volga_price(
                    df_d=self.df_d,
                    df_f=self.df_f,
                    p=1.0,
                    S_t=S,
                    sigma_atm=self.sigma_atm,
                    sigma_25C=self.sigma_25C,
                    sigma_25P=self.sigma_25P,
                    t=self.market.start,
                    delta_forward=self.market.delta_forward,
                    delta_premium=self.market.delta_premium,
                    base=365,
                    dK=0.0001
                    ) 
            else:
                raise ValueError("Metoda to nie BS ani VV")

            prices.append(price)
        return spots, np.array(prices)
    
    #funkcja poniżej dla danej funkcji wywołuje option_price_range, bierze od niej wartości i umieszcza je na wykresie
    #AI w poniższej funkcji przypomniało mi jak robi się wykresy poprawnie
    def plot_vs_spot(self,product_type:str,K: float, option_type: OptionType=OptionType.CALL, bound_type:BoundType=BoundType.LOWER_BOUND,payment_type:PaymentType= PaymentType.AT_MATURITY ):
        
        spots, bs = self.option_price_range(product_type, option_type, K,bound_type=bound_type,payment_type=payment_type,pricing_method= "bs",)
        x, vv = self.option_price_range(product_type, option_type, K, bound_type=bound_type,payment_type=payment_type, pricing_method="vv")
        
        if product_type=="OneTouch" or product_type=="NoTouch": #opcje onetouch i notouch są oddzielnie z racji bariery
            plt.figure()

            plt.plot(spots, bs, label="Black-Scholes")
            plt.plot(spots, vv, label="Vanna-Volga")

            plt.axvline(self.market.spot, linestyle="--", label="Spot")
            plt.axvline(K, linestyle="--",color="green", label="Bariera")
            
            plt.xlabel("FX Spot")
            plt.ylabel("Cena Opcji")

            if product_type=="OneTouch":
                plt.title(f"Cena opcji vs Spot dla {product_type} {bound_type.name} {payment_type.name}")
            else:
                plt.title(f"Cena opcji vs Spot dla {product_type} {bound_type.name}")
            plt.legend()
            plt.show()
        else:
            plt.figure()

            plt.plot(spots, bs, label="Black-Scholes")
            plt.plot(spots, vv, label="Vanna-Volga")

            plt.axvline(self.market.spot, linestyle="--", label="Spot")

            plt.xlabel("FX Spot")
            plt.ylabel("Cena Opcji")
        
            plt.title(f"Cena opcji vs Spot dla {product_type} {option_type.name}")
            plt.legend()
            plt.show()
```

```python
#Wywołaj tą komórkę, aby wyświetliła się wycena
tenor="1W" #dummy zmienne, aby nie wyświetlał sie błąd
typ_opcji="Vanilla"
bound_type="OneTouch"
B=4
call_put="Call"
payment_type="At touch"

tenor=input("Wybierz tenor: 1W, 2W, 1M, 2M, 3M, 6M, 9M, 1Y") #możliwe tenory
typ_opcji=input("Wybierz typ opcji: Vanilla, AoN, CoN, OneTouch, NoTouch") #możliwe typy opcji
if typ_opcji in ["OneTouch", "NoTouch"]: #dla one touchy jest up down zamiast call i put
    bound_type=input("Wybierz typ bariery: Up, Down")
    B=input("Wybierz poziom bariery:")
    B = float(B)
    if typ_opcji == "OneTouch":
        payment_type=input("Wybierz czas wypłaty: At touch, At maturity")
else: 
    call_put=input("Wybierz: Call lub Put")

engine = PricingEngine("market_data.xlsx", tenor=tenor,imply_pln=True) #załadowanie silnika dla danego tenoru

if call_put=="Call":   #zamiana inputów na wartości enum
    call_put=OptionType.CALL
else:
    call_put=OptionType.PUT
if bound_type=="Down":
    bound_type=BoundType.LOWER_BOUND
else:
    bound_type=BoundType.UPPER_BOUND
if payment_type=="At touch":
    payment_type=PaymentType.AT_TOUCH
else:
    payment_type=PaymentType.AT_MATURITY
    
match typ_opcji:     #case, aby dobrać dobry typ opcji 
    case "Vanilla":
        option_price = engine.vanilla_price( #wywołanie wyceny
        K=engine.strike(0.25, call_put),
        option_type=call_put,
        pricing_method="vv"
        )
    case "CoN":
        option_price = engine.con_price(
        K=engine.strike(0.25, call_put),
        option_type=call_put,
        pricing_method="vv"
        )
    case "AoN":
        option_price = engine.aon_price(
        K=engine.strike(0.25, call_put),
        option_type=call_put,
        pricing_method="vv"
)
    case "NoTouch":
        option_price = engine.no_touch_price(
        B=B,
        option_type=OptionType.CALL, #nie ma znaczenia, taka implementacja
        bound_type=bound_type,
        pricing_method="vv"
        ) 
    case "OneTouch":
        option_price = engine.one_touch_price(
        B=B,
        option_type=OptionType.CALL, #nie ma znaczenia, taka implementacja
        bound_type=bound_type,
        payment_type=payment_type,
        pricing_method="vv"
)
    case _:
        raise ValueError(f"Nieznany typ opcji: {product_type}")
print("Cena opcji wynosi: ",option_price)
if typ_opcji in ["OneTouch", "NoTouch"]: 
    #funkcja plot vs spot przyjmuje wyłącznie K, jednak dla opcji touch i no touch zamiast dodawać kolejną zmienną na B, 
    #można po prostu użyc istniejącego K
    K=B
else:
    K=engine.strike(0.25, call_put)

engine.plot_vs_spot(K=K,product_type=typ_opcji,option_type=call_put,bound_type=bound_type,payment_type=payment_type) #wywołanie wykresu
```
